# Series.

Pandas Series objects are generally created from the following data types.<br>
1.Scalar (Single Value)<br>
2.ndarray<br>
3.List<br>
4.List<br>

pd.Series(data,index,dtype)

```python
import pandas as pd
```

```python
# Creting from a scalar value.
x = pd.Series("string")
x
```

```python
# creating from an ndarray.
import numpy as np
arr = np.array([1,2,3,4,5])
series1 = pd.Series(arr)
print(series1)
```

```python
# creating from a list
lst = [1,2,3,4,5,6,7,8]
y = pd.Series(lst)
print(y)
```

```python
x = [1,2,5,2,4,33,2,3,5,6]
y = pd.Series(x)
y
```

```python
# Creating from an Dictionary.
# key of the dictionary become the index.
dict1 = {"a":1,"b":2,"c":3,"d":4}
d = pd.Series(dict1)
print(d)
```

# Reindexing.
An indexing method on pandas objects is *reindex*, which means to be create a new objects with values rearranged according to the new index.

```python
import pandas as pd
```

```python
obj = pd.Series([45,34,25,34],index=["d","c","b","a"])
obj
```

```python
t = pd.DataFrame([12,3,4,5,6,7])
t
```

```python
obj2 = obj.reindex({"a","b","c","d"})
obj2
```

```python
obj
```

```python
# A Series's index can be altered in place by assignment
obj.index = ["a","b","c","d"]
obj
```

with DataFrame,*reindex* can after the index,columns, or both.

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame(np.arange(9).reshape((3,3)),index=["a","c","b"],columns=["Y","Z","X"])
df
```

```python
df = df.reindex(index=["a","b","c"],columns=["X","Y","Z"])
df
```

```python
u = pd.DataFrame(np.arange(9).resize((3,3)),index=["a","c","b"],columns=["Y","Z","X"])
u
```

```python
df = pd.DataFrame(np.arange(9).reshape((3,3)))
df
```

# Indexing,Selection and filtering.

***Indexing with series***

Series indexing works similar to NumPy array indexing, except you can use the Series's index values(labels)instead of only integers.

```python
import pandas as pd
```

```python
ser1 = pd.Series([1,2,3,4])
print(ser1)
print()
ser1[0]
```

```python
s1 = pd.Series
```

```python
ser2 = pd.Series([5,6,7,8],index=["a","b","c","d"])
print(ser2)
print()
ser2["a"]
```

```python
ser2[3]  # Numerical also works.
```

```python
ser2[["a","c","d"]] # selecting a set of values.
```

```python
ser2[[0,1,3]]
```

```python
ser2[0::2]
```

While you can select data by label this way,the prefferd way to select index values is with the special loc operator

```python
obj1 = pd.Series([1,2,3],index=[2,0,1])
obj2 = pd.Series([1,2,3],index=["a","b","c"])
print(obj1)
```

```python
print(obj2)
```

```python
obj1[0]
```

```python
obj2[0]
```

The .loc operator is used to select data by label

```python
obj2.loc[[0,1,2]] # error because [0,1,2] are not the labels of obj2
```

```python
obj2.loc[["a","b"]]
```

The .iloc operator indexes only with integers and not by labels.

```python
print(obj1)
```

```python
obj1.iloc[0:2]    # integer based
```

```python
obj1.loc[2:0]  # label based
```

```python
print(obj2)
```

```python
obj2.iloc[2]   # Integer based
```

```python
obj2.loc[2] # label based
```

```python
# start and stop index are inclusive when slicing by labels (by using or not using .loc)
print(obj2)
```

```python
obj2["a":"c"]
```

```python
obj2.loc["a":"c"]
```

```python
obj2.iloc[0:2]
```
